```python
def rastrigin(x, y):
    return 20 + x**2 + y**2 - 10 * (np.cos(2 * np.pi * x) + np.cos(2 * np.pi * y))
```

```python
print(rastrigin(2,2))
```

```python
def himmelblau(x, y):
    return (x**2 + y - 11)**2 + (x + y**2 - 7)**2
```

```python
print(himmelblau(2,2))
```

```python
#plot rastrigin
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x1 = np.linspace(-5.12, 5.12, 400)
y1 = np.linspace(-5.12, 5.12, 400)
X1, Y1 = np.meshgrid(x1, y1)
Z1 = rastrigin(X1, Y1)

# Create plots
fig = plt.figure(figsize=(14, 6))

# 3D Surface Plot
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(X1, Y1, Z1, cmap='viridis')
ax1.set_title('Rastrigin Function - 3D Surface')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x, y)')

# 2D Contour Plot
ax2 = fig.add_subplot(1, 2, 2)
contour1 = ax2.contourf(X1, Y1, Z1, levels=50, cmap='viridis')
fig.colorbar(contour1, ax=ax2)
ax2.set_title('Rastrigin Function - 2D Contour')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

plt.tight_layout()
plt.show()
```
